### 1. Analyze the relationship between cholesterol levels and heart attack risk, considering additional factors such as BMI, blood pressure, triglyceride levels, and lifestyle habits. What patterns indicate a stronger correlation?

Data: https://drive.google.com/file/d/13ShLPdVJdjiCsyIuif4Ih45ZUA9Zu8Zz/view?usp=sharing

```python
#Importing the required libraries
import pandas as pd
import numpy as np
from datetime import datetime
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv('Heart_Attack_Male.csv')
df.head(5)
```

```python
# Summary Statistics of the dataset
df.describe()
```

```python
# Checkiing the datatype for each column
df.info()
```

```python
# Checking for duplicates
df.duplicated().sum()
```

```python
# To check relatinoships or correlation, categorical columns need conversion to numerical columns

# For nominal categorical columns
nominal_columns = df[['Heart Attack Likelihood', 'Diet Type','Exercise Induced Angina','Region', 'Urban/Rural', 'Family History of Heart Disease', 'Diabetes', 'Hypertension', 'ECG Results', 'Chest Pain Type']]


# Use Label enconder for nominal columns
encoder = LabelEncoder()
for col in nominal_columns:
    df[col] = encoder.fit_transform(df[col])
    
    
df.head(3)

```

```python
# Encoding for ordinal categorical columns
column_mappings = {
    'Smoking Status': {'Never': 0, 'Occasionally': 1, 'Regularly': 2},
    'Alcohol Consumption': {'Never': 0, 'Occasionally': 1, 'Regularly': 2},
    'Physical Activity Level': {'Sedentary': 0, 'Moderate': 1, 'High': 2},
    'Stress Level': {'Low': 0, 'Medium': 1, 'High': 2},
    'SES': {'Low': 0, 'Middle': 1, 'High': 2}
}

for col, mapping in column_mappings.items():
    df[col] = df[col].map(mapping)
        
df.head(3)
```

```python
""""Breaking blood pressure column into diastolic and systolic blood pressure"""

# first converting the data type to string so I can split
df['Blood Pressure (systolic/diastolic mmHg)'] = df['Blood Pressure (systolic/diastolic mmHg)'].astype(str)

# Spliting into Systolic and Diastolic
df[['Systolic_BP', 'Diastolic_BP']] = df['Blood Pressure (systolic/diastolic mmHg)'].str.split('/', expand=True)

# Converting to numeric format 
df['Systolic_BP'] = pd.to_numeric(df['Systolic_BP'], errors='coerce')
df['Diastolic_BP'] = pd.to_numeric(df['Diastolic_BP'], errors='coerce')

# Dropping the original 'Blood Pressure' column
df.drop(columns=['Blood Pressure (systolic/diastolic mmHg)'], inplace=True)

# dropping the gender column since all is male
df.drop('Gender', axis=1, inplace=True)
```

```python
"""Boxplot showing the relationship between Cholesterol level and Heart Attack Likelihood"""
# seting the fig size
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Heart Attack Likelihood'], y=df['Cholesterol Levels (mg/dL)'])

# Title and labels
plt.title('Cholesterol Levels vs. Heart Attack Likelihood')
plt.xlabel('Heart Attack Likelihood')
plt.ylabel('Cholesterol Levels (mg/dL)')

# Show the plot
plt.show()
```

The median cholesterol levels for both groups( Heart Attack Likelihood = and =1) appears to be same, same goes for the lowest and highest cholesterol levels for both groups, hence, cholesterol levels cannot be used as a single predictor for heart attack likelihood in this case.

```python
# finding the correlation between Heart attack likehood and other features
Correlation = df.corr()['Heart Attack Likelihood'].sort_values(ascending = False)
Correlation
```

Query 1 Answer: For the relationship between heart attack likelihood and the rest of the features, Resting heart rate has the strongest correlation with heart attack likelihood with a weak positve correlation 0.037163 while social economic status has a very weak negative correlation of -0.023178, meaning as people with higher socioeconomic status may have a slightly lower risk of heart attacks.

### 2. Visualize how profit is distributed across different categories using a pie chart, and explode the wedge with the highest percentage of profit across all categories.

Data: https://drive.google.com/file/d/1-cCEA9WxaLbjx8MwQJUfrQOfznnE_r5k/view?usp=sharing

```python
# Load the data
df_2 = pd.read_csv('Superstore (1).csv')
df_2.head(5)
```

```python
# Checking for duplicates
df_2.duplicated().sum()
```

```python
# Removing duplicates
df_2.drop_duplicates(inplace=True)
```

```python
# Checking the unique categories to ensure a pie plot will be apprpopriate
df_2['Category'].unique()
```

```python
"""Visualizing Profit Distribution across different categories"""

# Grouping profit by cateogy
profit_by_category = df_2.groupby('Category')['Profit'].sum()

# finding the category with the highest profit
highest_profit_category = profit_by_category.idxmax()

# Creating an explode for the category with the highest profit
explode = [0.1 if category == highest_profit_category else 0 for category in profit_by_category.index]

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(profit_by_category, labels=profit_by_category.index, autopct='%1.1f%%', explode=explode, startangle=140) 

# Adding title
plt.title('Percentage Profit Distribution Across Categories')
```

Query 2 Answer: Technology has the highest percentage profit of all the categories with 50.8% of the total profit.

### 3. Visualize the global sales trends of Nintendo games over time using a line plot. How have Nintendo's sales figures evolved from the 1980s to the 2000s?

Data: https://drive.google.com/file/d/1IwbdhYcqMEf4zZH0Hfgbev9pCrpvJiGk/view?usp=sharing

```python
# Load the data
df_3 = pd.read_csv('vgsales_no_headers.csv', header = None)
df_3.head(5)
```

```python
# Assign column names manually
df_3.columns = ["S/N", "Game", "Platform", "Year", "Category", "Company", 
              "Region_A_Sales", "Region_B_Sales", "Region_C_Sales", "Region_D_Sales", "Total_Global_Sales"]

df_3.head(3)
```

```python
# Check for duplicates in the dataset
df_3.duplicated().sum()
```

```python
# Checking the data formats
df_3.info()
```

```python
# Converting the year column to datetime format
df_3['Year'] = pd.to_datetime(df_3['Year'], format='%Y')
```

```python
# Group by Year and sum Global Sales
global_trend = df_3.groupby('Year')['Total_Global_Sales'].sum().reset_index()

# Plotting the global sales trend with a line chart
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Total_Global_Sales', data=global_trend, marker='o', linewidth=2)

# Labels and title
plt.xlabel("Year")
plt.ylabel("Total Global Sales (millions)")
plt.title("Global Game Sales Trend Over the Years")
plt.grid(True)
plt.show()
```

Query 3 Answer: The Nintendo global sales trend started very low in the 1980s, there were fluctuations with little growth till 1995 when the sales began to experience rapid growth. The exponential growth continued till  mid 2000s when here was a sharp decline in game sales and the downward trend continued till 2020.

### 4. In Saudi Arabia, the 5-day workweek starts on Sunday. Analyze the impact of the weekend and weekday quote dates on the quote price.

Data: https://drive.google.com/file/d/1vROMYOn9OUrIS7ChIIzkGpjeEdxm5Pe1/view?usp=sharing

```python
# Load the data
df_4 = pd.read_csv('luxury_goods_selling_transaction - quote_sell_dataset - luxury_goods_selling_transaction - quote_sell_dataset.csv')
df_4.head(5)
```

```python
# Checking the info to confirm 'Quote-date is in datetime format'
df_4.info()
```

```python
# convert Quote date to date time format
df_4['Quote Date'] = pd.to_datetime(df_4['Quote Date'])

# Create a column to capture days of the week
df_4['Day_of_Week'] = df_4['Quote Date'].dt.day_name()

# mark days as weekday or weekend
df_4['Day_type'] = df_4['Day_of_Week'].apply(lambda x: 'Weekend' if x in [4, 5] else 'Weekday')
df_4.head()
```

```python
df_4['Day_of_Week'].unique()
```

Surprising, Sundays are not represented in the days of the week and as such, there are just 4 days in the weekday day type and 2 days for weekends

```python
# Extract day of the week (0 = Monday, 6 = Sunday)
df_4['Day_of_Week'] = df_4['Quote Date'].dt.dayofweek

# Mark Friday (4) and Saturday (5) as 'Weekend'
df_4['Day_type'] = df_4['Day_of_Week'].apply(lambda x: 'Weekend' if x in [4, 5] else 'Weekday')
df_4.head()
```

```python
df_4['Day_type'].unique()
```

```python
# Group by Day Type and calculate summary stats for quote price
quote_summary = df_4.groupby('Day_type')['Quote Price'].agg(['mean', 'median', 'sum', 'count']).reset_index()
quote_summary
```

```python
total_prices = df_4.groupby('Day_type')['Quote Price'].sum().reset_index()

# Pie chart showing total prices by day type
plt.figure(figsize=(8, 5))

labels = total_prices['Day_type']
values = total_prices['Quote Price']

plt.pie(values, labels=labels, autopct='%1.1f%%', colors=['skyblue', 'orange'], startangle=140)
      
# Title
plt.title('Total Quote Price Comparison: Weekday vs Weekend')

# Show plot
plt.show()
```

```python
"""Boxplot showing the relationship between Day type and Quote Price"""
# seting the fig size
plt.figure(figsize=(8, 5))
sns.boxplot(x=df_4['Day_type'], y=df_4['Quote Price'])

# Title and labels
plt.title('Quote Price vs. Day Type')
plt.xlabel('Day Type')
plt.ylabel('Quote Price')

# Show the plot
plt.show()
```

Query 4 Answer: From the analysis, the median quote price for weekends is higher than for weekdays which might indicate that quote prices tend to be slightly elevated on weekends, possibly due to increased demand. .
